```python
# 
import os
import cv2 as cv
from tqdm import tqdm
```

```python
! rm -rf ../out/T000001/images/*.png*
```

```python
pi = "../data/images/"
pb = "../data/imagesb/"
po = "../out/T000001/images/"

# 7, 294/297
n = 7
m = 297

for i in tqdm(os.listdir(pi)[:]):
    img = cv.imread(f"{pi}{i}")
    img0 = cv.imread(f"{pb}{i}")
    
    img[0:n,m-n:m] = img0[0:n,m-n:m]
    img[m-n:m,m-n:m] = img0[m-n:m,m-n:m]
    
    cv.imwrite(f"{po}{i}", img)
```
